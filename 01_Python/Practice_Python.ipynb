```python
# Largest Element in List

def largest(arr):
  try:
    arr.sort()
    return arr[-1]
  except:
    print("input is not a valid list or may be a empty list passed")

def man_largest(arr):
  try:
    largest = arr[0]

    for num in arr[1:]:
        if num > largest:
            largest = num

    return largest
  except:
        print("input is not a valid list or may be a empty list passed")
a=[6,5,8,4,17,12,3]
b = []
print(largest(a))
print(man_largest(a))

print(largest(b))
print(man_largest(b))
```

```python
# WAP to find smallest and largest number in list  (without min-max function)
def small_large(arr):
  small = large = arr[0]
  for num in arr:
    if num > large:
      large = num
    elif num < small:
      small = num

  return f"small = {small} & large = {large}"

d = [3,4,5,17,17, 27, 4,24, 27, 7 , 9, 2,11]
print(small_large(a))
print(small_large(c))
print(small_large(d))

```
