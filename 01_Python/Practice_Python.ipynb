```python
# Largest Element in List

def largest(arr):
  try:
    arr.sort()
    return arr[-1]
  except:
    print("input is not a valid list or may be a empty list passed")

def man_largest(arr):
  try:
    largest = arr[0]

    for num in arr[1:]:
        if num > largest:
            largest = num

    return largest
  except:
        print("input is not a valid list or may be a empty list passed")
a=[6,5,8,4,17,12,3]
b = []
print(largest(a))
print(man_largest(a))

print(largest(b))
print(man_largest(b))
```

```python
def largest(arr):
    if arr is None:
        raise TypeError("arr cannot be None")

    if len(arr) == 0:
        raise ValueError("Empty list")

    largest = arr[0]

    for num in arr[1:]:
        if num > largest:
            largest = num

    return largest
```
